```python
df_total = df.sort_values(by = 'Total', ascending=False )
df_total = df_total.head(20)
```

```python
top20_total = px.treemap(
    df_total,
    path=['Name', 'Total'],
    title='Top 20 Pokemon',
    values='Total',
    height=600
)

top20_total.show()
```

Top 20 pokemon based on total stats

```python
fig = px.box(df[df['Total'] < 600],
             x='Total',
             orientation='h',
             title='Pokemon with total stats less than 600')
fig.show()
```
